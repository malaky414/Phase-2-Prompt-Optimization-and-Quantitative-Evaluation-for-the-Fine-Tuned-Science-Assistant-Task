```python
from peft import PeftModel
from transformers import AutoModelForCausalLM, AutoTokenizer

base_model = AutoModelForCausalLM.from_pretrained("distilgpt2")

model = PeftModel.from_pretrained(
    base_model,
    "/content/drive/MyDrive/lora_model"
)

tokenizer = AutoTokenizer.from_pretrained("/content/drive/MyDrive/lora_model")
```

```python
def prompt_direct(q):
    return f"{q}"
```

```python
def prompt_structured(q):
    return f"Question: {q}\nAnswer:"
```

```python
def prompt_explain(q):
    return f"Answer the question and explain briefly:\nQuestion: {q}\nAnswer:"
```

```python
def generate(prompt, temp=0.7):
    inputs = tokenizer(prompt, return_tensors="pt")

    outputs = model.generate(
        **inputs,
        max_new_tokens=50,
        temperature=temp,
        do_sample=True
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
results = []

for example in eval_data:
    print(type(example))
    q = example["question"]
    real = example["correct_answer"]

    results.append({
        "question": q,
        "real": real,
        "direct": generate(prompt_direct(q)),
        "structured": generate(prompt_structured(q)),
        "explain": generate(prompt_explain(q))
    })
```

```python
import json

with open("prompt_results.json", "w") as f:
    json.dump(results, f, indent=4)
```

```python
temp_results = []

for example in list(eval_data)[:10]:
    print(type(example))
    q = example["question"]

    temp_results.append({
        "question": q,
        "low_temp": generate(prompt_structured(q), temp=0.3),
        "high_temp": generate(prompt_structured(q), temp=0.9)
    })
```

```python
import evaluate

bleu = evaluate.load("bleu")
rouge = evaluate.load("rouge")

predictions = [r["structured"] for r in results]
references = [[r["real"]] for r in results]

bleu_score = bleu.compute(predictions=predictions, references=references)
rouge_score = rouge.compute(predictions=predictions, references=[r["real"] for r in results])

print("BLEU:", bleu_score)
print("ROUGE:", rouge_score)
```

```python
base_model = AutoModelForCausalLM.from_pretrained("distilgpt2")

def generate_base(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = base_model.generate(**inputs, max_new_tokens=50)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
comparison = []

for example in list(eval_data)[:5]:
    q = example["question"]

    comparison.append({
        "question": q,
        "base": generate_base(prompt_structured(q)),
        "fine_tuned": generate(prompt_structured(q))
    })
```
